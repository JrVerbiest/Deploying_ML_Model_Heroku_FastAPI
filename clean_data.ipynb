# Clean data

```python
import pandas as pd
```

```python
PATHDATA = './data/census.csv'
PATHCLEANDATA = './data/clean_census.csv'
```

```python
# read data, using method skipinitialspace to remove spaces
df = pd.read_csv(PATHDATA, skipinitialspace = True)
df.head()
```

```python
# do some extra cleaning
df.replace("?", None, inplace=True)
df.dropna(inplace=True)
```

```python
# save cleaned data
df.to_csv(PATHCLEANDATA, index=False)
```
